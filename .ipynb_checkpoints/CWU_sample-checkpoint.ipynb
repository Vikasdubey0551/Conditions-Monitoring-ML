```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
import numpy as np
import h5py
from tabulate import tabulate
```

```python
# Sample frequency
f_sample = 12e3
# Axis rotation frequency
f_rotation = 28.68
bpfo_multiplier = 3.5848
f_bpfo = f_rotation * bpfo_multiplier
# print("Expected BPFO first frequency " + str(f_bpfo) + " Hz")
f_bpfo_harmonics = f_bpfo*(np.arange(1, 20, 1))
# print( "Expected BPFO harmonic frequencies " + str(f_bpfo_harmonics) + " Hz")
```

```python

hf = h5py.File("x_baseline.h5", "r")
x_basestr = list(hf.keys())
x_str = x_basestr[0]
ds_baseline = hf[x_str]
x_baseline = np.array(hf.get(x_str))
hf = h5py.File("x_fault.h5", "r")
x_faultstr = list(hf.keys())
x_str = x_faultstr[0]
x_fault = np.array(hf.get(x_str))
t = np.linspace(0, 1, int(f_sample))
print(x_baseline.shape)
```

```python
# Data processing in time domain
def time_calcer(vibration_array):
    from tabulate import tabulate
    x = vibration_array
    Z_rms = np.int(100 * np.sqrt(np.sum(x ** 2) / len(x))) / 100
    Z_peak = np.int(100 * np.amax([np.abs(np.amax(x)), np.abs(np.amin(x))])) / 100
    Z_pk2pk = np.int(100 * np.abs(np.amax(x) - np.amin(x))) / 100
    Z_crest = np.int(100 * Z_peak / Z_rms) / 100
    headers = ['Item', 'Value']
    table_time = [['Item', 'Value'], ['RMS', Z_rms], ['Peak', Z_peak], ['Pk-Pk', Z_pk2pk], ['Crest', Z_crest]]
    # print(tabulate(table_time, headers, tablefmt="psql"))
    # print(tabulate(table_time, tablefmt="psql"))
    return table_time
plt2.subplot(2, 2, 1)
plt2.figure(1, figsize=(5, 5))
for j in range(x_baseline.shape[0]):
    x = x_baseline[0, :]
    if j == 0:
        table_t = time_calcer(x)
    plt2.plot(t, x)
plt2.grid('major')
plt2.title('Baseline data set')
plt2.xlabel('Time (sec)')
plt2.ylabel('Amplitude (g)')
plt2.subplot(2, 2, 3)
plt2.axis("off")   # turns off axes
plt2.axis("tight")  # gets rid of white border
plt2.table(table_t, loc='center')
# Fault data processing in time domain
plt2.subplot(2, 2, 2)
plt2.figure(1, figsize=(5, 5))
for j in range(x_fault.shape[0]):
    x = x_fault[0, :]
    if j == 0:
        table_t = time_calcer(x)
    plt2.plot(t, x)
plt2.grid('major')
plt2.title('Fault data set')
plt2.xlabel('Time (sec)')
plt2.ylabel('Amplitude (g)')
plt2.subplot(2, 2, 4)
plt2.axis("off")   # turns off axes
plt2.axis("tight")  # gets rid of white border
plt2.table(table_t, loc='center')
plt2.show()
```

```python
# Baseline data processing in frequency domain
def peak_finder(freq, fft_amp):
    from tabulate import tabulate
    lastval = fft_amp[0]
    lastindex = 0
    dirpos = True
    peaks_amp = []
    peaks_freq = []
    for i in range(fft_amp.size):
        currentval = fft_amp[i]
        if dirpos and currentval < lastval:
            peaks_amp.append(fft_amp[lastindex])
            peaks_freq.append(freq[lastindex])
            dirpos = False
        if currentval > lastval:
            dirpos = True
        lastval = currentval
        lastindex = i
    peaks_amp = np.asarray(peaks_amp)
    peaks_freq = np.asarray(peaks_freq)
    peaks5 = np.sort(peaks_amp)[::-1]
    peaks5 = peaks5[0:5]
    peaks_table = np.asarray([1, 2])
    Z_freqs = ['Freq.(Hz)']
    Z_amps = ['Amplitude (Peak)']
    for i in range(peaks5.size):
        j = np.where(peaks_amp == peaks5[i])
        index = np.int(j[0])
        a = [peaks_amp[index], peaks_freq[index]]
        Z_amps.append(np.int(100 * a[0]) / 100)
        Z_freqs.append(np.int(10 * a[1]) / 10)
    headers_f = ['Freq.(Hz)', 'Amplitude (Peak)']
    table_freq = [Z_freqs, Z_amps]
    table_freq = np.transpose(np.asarray(table_freq))
    # print(tabulate(table_freq, headers_f, tablefmt="psql"))
    # table_freq = np.append(headers_f,table_freq, axis=0)
    return table_freq
def fft_amp(time_array, vib_array):
    # This function returns the fft absolute amplitude according to time and vibration arrays
    x0 = np.hanning(len(vib_array))
    vib_array = x0 * vib_array
    amp = np.fft.fft(vib_array) / (len(vib_array) / 2)  # Normalize amplitude
    amp = 2 * abs(amp[range(int(len(vib_array) / 2))])
    fs1 = 1 / (time_array[1] - time_array[0])
    freq = np.linspace(0, fs1 / 2, np.int(amp.size))
    return freq, amp
plt.subplot(2, 2, 1)
plt.figure(1, figsize=(5, 5))
for j in range(x_baseline.shape[0]):
    x = x_baseline[0, :]
    freq, amp = fft_amp(t, x)
    if j ==0 :
        table_f = peak_finder(freq, amp)
    plt.plot(freq, amp)
plt.grid('major')
plt.title('Baseline data set')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (g)')
plt.subplot(2, 2, 3)
plt.axis("off")   # turns off axes
plt.axis("tight")  # gets rid of white border
plt.table(table_f, loc='center')

# Fault data processing in frequency domain
plt.subplot(2, 2, 2)
plt.figure(1, figsize=(5, 5))
for j in range(x_fault.shape[0]):
    x = x_fault[0, :]
    freq, amp = fft_amp(t, x)
    if j ==0 :
        table_f = peak_finder(freq, amp)
    plt.plot(freq, amp)
plt.grid('major')
plt.title('Fault data set')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (g)')
plt.subplot(2, 2, 4)
plt.axis("off")   # turns off axes
plt.axis("tight")  # gets rid of white border
plt.table(table_f, loc='center')
plt2.show()
```
