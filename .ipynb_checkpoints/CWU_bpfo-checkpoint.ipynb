```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
import pandas as pd
import seaborn as sb
import sklearn as sk
```

```python
# Sample frequency
f_sample = 12e3
# Rotational speed of the shaft (Hz)
f_rotation = 28.68
# Bearing Pass Frequency Outer Race Multiplier
bpfo_multiplier = 3.5848
f_bpfo = f_rotation * bpfo_multiplier
# print("Expected BPFO first frequency " + str(f_bpfo) + " Hz")
f_bpfo_harmonics = f_bpfo*(np.arange(1, 21, 1))
print( "Expected BPFO harmonic frequencies " + str(f_bpfo_harmonics) + " Hz")
```

```python
# Loading baseline and fault data from data files
hf = h5py.File("x_baseline.h5", "r")
x_basestr = list(hf.keys())
x_str = x_basestr[0]
ds_baseline = hf[x_str]
x_baseline = np.array(hf.get(x_str))

hf = h5py.File("x_fault.h5", "r")
x_faultstr = list(hf.keys())
x_str = x_faultstr[0]
x_fault = np.array(hf.get(x_str))
```

```python
# Plotting time signals versus each other
# Baseline data processing in time domain
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
for j in range(x_baseline.shape[0]):
    x = x_baseline[0, :]
    plt.plot(t, x)
plt.grid('major')
plt.title('Baseline data set')
#plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (g)')
# Fault data processing in time domain
plt.subplot(2, 1, 2)
for j in range(x_fault.shape[0]):
    x = x_fault[0, :]
    plt.plot(t, x)
plt.grid('major')
plt.title('Fault data set')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (g)')
plt.show()

```

```python
# FFT calculation function
def fft_amp(time_array, vib_array):
    # This function returns the fft absolute amplitude according to time and vibration arrays
    x0 = np.hanning(len(vib_array))
    vib_array = x0 * vib_array
    amp = np.fft.fft(vib_array) / (len(vib_array) / 2)  # Normalize amplitude
    amp = 2 * abs(amp[range(int(len(vib_array) / 2))])
    fs1 = 1 / (time_array[1] - time_array[0])
    freq = np.linspace(0, fs1 / 2, np.int(amp.size))
    return freq, amp
```

```python
# Baseline data processing in frequency domain
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
for j in range(x_baseline.shape[0]):
    x = x_baseline[0, :]
    freq, amp = fft_amp(t, x)
    plt.plot(freq, amp)
plt.grid('major')
plt.title('Baseline data set')
# plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (g)')

# Fault data processing in frequency domain
plt.subplot(2, 1, 2)
for j in range(x_fault.shape[0]):
    x = x_fault[0, :]
    freq, amp = fft_amp(t, x)
    plt.plot(freq, amp)
plt.grid('major')
plt.title('Fault data set')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (g)')
plt.show()
```

```python
# This function finds the nearest frequency values to BPFO harmonics
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx], idx
```

```python
# Finding the nearest indices of FFT to BPFO harmonics
t = np.linspace(0, 1, int(f_sample))
x = x_baseline[0, :]
freq, amp = fft_amp(t, x)
indices = []
freqs_bpfo = []
for j in range(len(f_bpfo_harmonics)):
    a, b = find_nearest(freq, f_bpfo_harmonics[j])
    indices.append(b)
    freqs_bpfo.append(a)
```

```python
# Baseline feature extraction for BPFO harmonics
features = np.zeros((x_baseline.shape[0] + x_fault.shape[0], 20))
for j in range(x_baseline.shape[0]):
    x = x_baseline[j, :]
    freq, amp = fft_amp(t, x)
    for i in range(len(f_bpfo_harmonics)):
        features[j, i] = amp[indices[i]]

# Fault feature extraction for BPFO harmonics
for j in range(x_fault.shape[0]):
    x = x_fault[j, :]
    freq, amp = fft_amp(t, x)
    for i in range(len(f_bpfo_harmonics)):
        features[j + (x_baseline.shape[0]), i] = amp[indices[i]]

# Creating target values 0 = OK , 1 = Not OK
df_col = (x_baseline.shape[0]) + (x_fault.shape[0])
target = np.zeros(df_col)
for j in range(df_col):
    if j > 39:
        target[j] = 1
```

```python
# Creating data frame for machine learning 
Col_names = []
for j in range(len(f_bpfo_harmonics)):
    str1 = str(j+1) + 'XBPFO'
    Col_names.append(str1)
my_df = pd.DataFrame(features, columns=Col_names)
my_df.describe()
num_var = my_df
```

```python
# Detection of underlying correlation of features
corr = num_var.corr('pearson')  # 'kendall', 'spearman', 'pearson'
# print('Correlation', corr)
plt.figure(figsize=(11,11))
sb.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, vmin=-1, vmax=1)
plt.show()
```

```python
# From above we see correlation between BPFO harmonics: 4,5,7,14
my_df1 = my_df[['4XBPFO', '5XBPFO', '7XBPFO', '14XBPFO']]
pd.plotting.scatter_matrix(my_df1, c=target, figsize=[11, 11], s=150)
plt.show()
```

```python
# Application of KNN Classifier to implement machine learning in detection of OK/Fault (0,1) from BPFO harmonics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42, stratify=target)
knn = KNeighborsClassifier(n_neighbors=7, metric='minkowski', p=2)
knn.fit(x_train, y_train)
y_predict_knn = knn.predict(x_test)
print('KNN Score for the prediction of vibration values =', int(100*knn.score(x_test, y_test))/100)
```

```python
# Comparison of prediction vs. real test data
x_plot = np.arange(len(y_predict_knn))
plt.plot(x_plot+1, y_test, 'b')
plt.plot(x_plot+1, y_predict_knn, 'ro')
plt.xlabel('Test number')
plt.ylabel('Prediction (0 Baseline, 1 Fault)')
plt.legend(['Test Data', 'Prediction'])
plt.grid('major')
plt.show()
```
